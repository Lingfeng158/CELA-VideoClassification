```python
import os
import subprocess
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.random import rand
import numpy as np
```

```python
dataDir='/Users/lingfengli/Desktop/SIR/cela-tf-starter/ResizeBinnedToClips6F/Training'
dirList=[]
```

```python
categories=os.listdir(dataDir)
if('.DS_Store' in categories):
    categories.remove('.DS_Store')
for category in categories:
    subjects=os.listdir(os.path.join(dataDir, category))
    if('.DS_Store' in subjects):
        subjects.remove('.DS_Store')
    for subject in subjects:
        clips=os.listdir(os.path.join(dataDir, category, subject))
        if('.DS_Store' in clips):
            clips.remove('.DS_Store')
        for clip in clips:
            dirList.append(os.listdir(os.path.join(dataDir, category, subject, clip)))
```
